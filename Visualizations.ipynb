```python
#2d kde plots
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(1)
numerical_1 =np.random.randn(100)
np.random.seed(2)
numerical_2 = np.random.randn(100)

fig,ax = plt.subplots(figsize=(8,8))
sns.kdeplot(data=numerical_1,
           data2=numerical_2,
           ax=ax,
           shade=True,
           color="blue",
           bw=1,
           legend=True,
           shade_lowest=True,
           cbar=True)
plt.show()
```

```python
#heat map
import pandas as pd
df = pd.DataFrame(dict(categorical_1=['apple', 'banana', 'grapes',
                                      'apple', 'banana', 'grapes',
                                      'apple', 'banana', 'grapes'], 
                  categorical_2=['A','A','A','B','B','B','C','C','C'], 
                  value=[10,2,5,7,3,15,1,6,8]))
pivot_table = df.pivot("categorical_1", "categorical_2", "value")
```

```python
print(pivot_table)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.heatmap(data=pivot_table,
           cmap=sns.color_palette("Blues"),
           ax=ax)
plt.show()
```

```python
#overlay plots
#to plot multiple variables on same plot
fig, ax1 = plt.subplots(figsize=(8,8))
ax2=ax1.twinx()
sns.lineplot(x=['A','B','C','D'], 
             y=[4,2,5,3],
             color='r',
             ax=ax1)
sns.lineplot(x=['A','B','C','D'], 
             y=[1,6,2,4], 
             color='b',
             ax=ax2)    
ax.legend(['alpha', 'beta'], facecolor='w')
plt.show()
```

```python
#combine bar chart and line plot with same x-axis but different y-axis
sns.set(style="white", rc={"lines.linewidth":3})

fig,ax1 = plt.subplots(figsize=(8,8))
ax2=ax1.twinx()

sns.barplot(x=['A', 'B', 'C', 'D'],
           y=[200,100,500,50],
           color='#004488',
           ax=ax1)
sns.lineplot(x=['A', 'B', 'C', 'D'],
            y=[2,5,4,7],
            color='r',
            ax=ax2,
            marker="o")
plt.show()
sns.set()
```

```python
#combining barplot into one plot
categorical_1=['A','B','C','D']
colors=['red','blue','green','orange']
numerical=[[6,7,8,2],
          [4,3,8,9],
          [2,1,4,5],
          [10,5,2,3]]

number_groups = len(categorical_1)
bin_width = 1/(number_groups+1)  #remember +1

fig,ax= plt.subplots(figsize=(8,8))

for i in range(number_groups):
    ax.bar(x=np.arange(len(categorical_1))+ i*bin_width,
          height=numerical[i],
          width=bin_width,
          color=colors[i],
          align='center')
ax.set_xticks(np.arange(len(categorical_1))+number_groups/(2*(number_groups+1)))

# number_groups/(2*(number_groups+1)): offset of xticklabel

ax.set_xticklabels(categorical_1)
ax.legend(categorical_1, facecolor='w')

plt.show()
```

```python
# partition into several subplots
import matplotlib
fig = plt.figure(figsize=(8,8))
gs = matplotlib.gridspec.GridSpec(nrows=3, 
                       ncols=3, 
                       figure=fig, 
                       width_ratios= [1, 1, 1],
                       height_ratios=[1, 1, 1],
                       wspace=0.3,
                       hspace=0.3)
ax1=fig.add_subplot(gs[0,0])
ax1.text(0.5,0.5,'ax1:gs[0,0]', fontsize=12, fontweight="bold", va="center", ha="center")
ax2=fig.add_subplot(gs[0,1:3])
ax2.text(0.5,0.5, 'ax2:gs[0,1:3]', fontsize=12, fontweight="bold", va="center", ha="center")
ax3=fig.add_subplot(gs[1:3,0:2])
ax3.text(0.5,0.5, 'ax3:gs[1:3,0:2]', fontsize=12, fontweight="bold", va="center", ha="center")
ax4=fig.add_subplot(gs[1:3,2])
ax4.text(0.5, 0.5, 'ax4:gs[1:3,0:2]', fontsize=12, fontweight="bold", va="center", ha="center")

plt.show()
```
